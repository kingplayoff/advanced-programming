# SOLUTION FOR ONE PLUS

https://leetcode.com/problems/plus-one/description/

**Solution 1:Using while loop **

```python
digits = [4,3,2,1]

def plusOne(digits):
    i = len(digits) - 1

    while i >= 0 and digits[i] == 9:
        digits[i] = 0
        i -= 1

    if i < 0:
        return [1] + digits

    digits[i] += 1
    return digits


print(plusOne(digits))
```

When executing, it will print -> [4,3,2,2]

**Solution 2: Convert to integer **

```python
digits = [1,2,3]
def plusOne(digits):
    num = int("".join(map(str, digits)))
    num += 1
    return list(map(int, str(num)))
print(plusOne(digits))
```

When executing, it will print -> [1,2,4]

**Solution 3:Loop from right**

```python
digits = [9]
def plusOne(digits):
    for i in range(len(digits)-1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits
        digits[i] = 0

    return [1] + digits
print(plusOne(digits))
```

When executing, it will print -> [1,0]
